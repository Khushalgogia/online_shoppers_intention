## Checking Bounce Rate and Exit rate (EDA)

```python
def bounce(x):
    if x > 0.05 and x <= 0.1:
        return '0.05 - 0.1'
    elif x > 0.1 and x <= 0.15:
        return '0.10 - 0.15'
    elif x > 0.15 and x <= 0.2:
        return '0.15 - 0.20'
    elif x >=0 and x <= 0.05:
        return '0 - 0.05'
    else:
        return '>0.20'
```

```python
pf = df1.copy()
```

```python
pf['bounce_rate'] = pf['BounceRates in %'].map(lambda x:bounce(x))
```

```python
pf['bounce_rate'].value_counts()
```

```python
pf['exit_rate'] = pf['ExitRates in %'].map(lambda x:bounce(x))
```

```python
def duration(x):
    if x >= 0 and x <= 500:
        return '0-500'
    elif x > 500 and x <= 1000:
        return '500-1000'
    elif x > 1000 and x <= 2500:
        return '1000-2500'
    elif x > 2500 and x <= 5000:
        return '2500-5000'
    elif x > 5000:
        return '>5000'
    else:
        return 'others'
```

```python
pf['prod_duration'] = pf['ProductRelated_Duration'].map(lambda x:duration(x))
```

```python
def admin_duration(x):
    if x >= 0 and x <= 60:
        return '0-60'
    elif x > 60 and x <= 300:
        return '60-300'
    elif x > 300 and x <= 600:
        return '300-600'
    elif x > 600 and x <= 1400:
        return '600-1400'
    elif x > 1400:
        return '>1400'
    else:
        return 'others'
```

## Splitting the data

```python
from sklearn.model_selection import train_test_split
xTrain,xTest,yTrain,yTest = train_test_split(df_scaled.iloc[:,:-1],df_scaled['Revenue__True'],test_size = 0.30,random_state= 101)
```

```python
print(xTrain.shape)
print(xTest.shape)
print(yTrain.shape)
print(yTest.shape)
```

## Checking Imbalance Dataset

```python
# Month__Jul ,Month__June ,Month__Oct, Month__Sep ,VisitorType__Other ,Revenue__True
# These columns are highly imbalanced data
```

```python
for i in list(df_scaled.columns):
    print(df_scaled[i].value_counts())
    print('#############'*5)
    
```

```python
from imblearn.over_sampling import RandomOverSampler
rs = RandomOverSampler()
xTrain,yTrain = rs.fit_resample(xTrain,yTrain)
```

```python
print(xTrain.shape)
print(yTrain.shape)
```

## Building Models

```python
from sklearn.metrics import accuracy_score
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
md1 = lr.fit(xTrain,yTrain)
pred1 = md1.predict(xTest)
acc1 = accuracy_score(pred1,yTest)

# Decision Tree
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
md2 = dt.fit(xTrain,yTrain)
pred2 = md2.predict(xTest)
acc2 = accuracy_score(pred2,yTest)

# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
md3 = rf.fit(xTrain,yTrain)
pred3 = md3.predict(xTest)
acc3 = accuracy_score(pred3,yTest)

# Bagging Classifier
from sklearn.ensemble import BaggingClassifier

bg = BaggingClassifier()
md4 = bg.fit(xTrain,yTrain)
pred4 = md4.predict(xTest)
acc4 = accuracy_score(pred4,yTest)

#SVM
from sklearn.svm import SVC

svm = SVC()
md5 = svm.fit(xTrain,yTrain)
pred5 = md5.predict(xTest)
acc5 = accuracy_score(pred5,yTest)

#KNN
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
md6 = knn.fit(xTrain,yTrain)
pred6 = md6.predict(xTest)
acc6 = accuracy_score(pred6,yTest)

# Voting Classifier
from sklearn.ensemble import VotingClassifier


vc = VotingClassifier(estimators = [('lr',lr),('dt',dt),('rf',rf),('bg',bg),('svm',svm),('knn',knn)])
md7 = vc.fit(xTrain,yTrain)
pred7 = md7.predict(xTest)
acc7 = accuracy_score(pred7,yTest)
```

```python
#Train score
ac1 = accuracy_score(md1.predict(xTrain),yTrain)
ac2 = accuracy_score(md2.predict(xTrain),yTrain)
ac3 = accuracy_score(md3.predict(xTrain),yTrain)
ac4 = accuracy_score(md4.predict(xTrain),yTrain)
ac5 = accuracy_score(md5.predict(xTrain),yTrain)
ac6 = accuracy_score(md6.predict(xTrain),yTrain)
ac7 = accuracy_score(md7.predict(xTrain),yTrain)
```

```python
l1 = ['Logistic Regression', 'Decision Tree', 'Random Forest Classifier' , 'Bagging Classifier', 'SVM', 'KNN', 'Voting Classifier']
scorecard = pd.DataFrame(data = {"Models": l1,'Train_score':[ac1,ac2,ac3,ac4,ac5,ac6,ac7],'Test_score': [acc1,acc2,acc3,acc4,acc5,acc6,acc7]})
scorecard.set_index('Models',inplace = True)
scorecard
```
